# Q1

```python
# 응답 설문 문항
question = ['Q2', 'Q3', 'Q4', 'Q5', 'Q6_1', 'Q6_2', 'Q6_3', 'Q6_4', 'Q6_5', 'Q6_6',
       'Q6_7', 'Q6_8', 'Q6_9', 'Q6_10', 'Q6_11', 'Q6_12', 'Q7_1', 'Q7_2',
       'Q7_3', 'Q7_4', 'Q7_5', 'Q7_6', 'Q7_7', 'Q8', 'Q9', 'Q10_1', 'Q10_2',
       'Q10_3']
```

#### 1번째 풀이

```python
# 리스트 초기화
answer = []

answer1 = [ques for ques in question if not '_' in ques]

# 출력
answer1
```

#### 2번째 풀이

```python
# 리스트 초기화
answer2 = []

for ques in question:
    if not '_' in ques:
        answer.append(ques)

# 출력
answer2
```

#### 3번째 풀이

```python
import re

# 리스트 초기화
answer3 = []

pattern = re.compile(r'^[A-Za-z0-9]+$')
for ques in question:
    if pattern.match(ques):
        answer3.append(ques)
        
# 출력
answer3
```

# Q2

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

#### 2011년부터의 데이터

```python
df_from_2011 = df[df['Year']>=2011]
df_from_2011.head()
```

```python
# Null 값 확인
df_from_2011.info()
```

#### 이상치값 확인하기
평균에 이상치 값이 큰 영향이 있어서 확인만,,,

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
sns.boxplot(data=df_from_2011, y='Spending_USD')
plt.title('Boxplot of Spending_USD')

plt.subplot(1, 2, 2)
sns.boxplot(data=df_from_2011, y='Life_Expectancy')
plt.title('Boxplot of Life_Expectancy')

plt.tight_layout()
plt.show()
```

```python
df_group_mean = df_from_2011.groupby(['Year', 'Country']).mean()
df_group_mean
```

```python
df_group_mean.unstack()['Life_Expectancy']
```
